```python
data.diagnosis = [1 if each== "M" else 0 for each in data.diagnosis]
y = data.diagnosis.values
x_data = data.drop(["diagnosis"],axis=1)
```

```python
#Normalization
```

```python
x = (x_data-np.min(x_data)) / (np.max(x_data)-np.min(x_data))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 1) #n_neighbors = k
knn.fit(x_train.values,y_train)
prediction = knn.predict(x_test.values)
```

```python
print("{} nÄ±n score:{} ".format(3,knn.score(x_test.values,y_test)))
```

# Find K value

```python
score_list = []
for each in range(1,15):
    knn2 = KNeighborsClassifier(n_neighbors = each)
    knn2.fit(x_train.values,y_train)
    score_list.append(knn2.score(x_test.values,y_test))

plt.plot(range(1,15),score_list)
plt.xlabel("K Values")
plt.ylabel("Accuracy")
plt.show()
```
